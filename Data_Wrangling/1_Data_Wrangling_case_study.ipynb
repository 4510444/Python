# Data Normalization

- Uniform the Data
- they have same impact
- Aik machli samundar me aur aik jar me
- Also for computational reasons

```python
ks1.head()
```

```python
ks2 = ks1[["age in days","fare"]]
ks2.head()
```

-  The above data is really in wide range and we need to naormalize and hard to compare
-  Normalization change the values to the range(0 to -1) now both the variable have significant influience on our models

## Methods of Normalization

- Simple feature scaling
    x(new) = x(old)/(xmax)
- Min-Max Method
- Z-score(Standard score) -3 to +3
- Log Transformation

```python
# simple feature scaling
ks2["fare"]=ks2["fare"]/ks2["fare"].max()
ks2["age in days"]=ks2["age in days"]/ks2["age in days"].max()
ks2.head()
```

```python
# Min-Max Method
ks2["fare"]= (ks2["fare"]-ks2["fare"].min())/(ks2["fare"].max()-ks2["fare"].min())
ks2.head()
```

```python
# Z-score (Standard Score) Method
ks2["fare"]= (ks2["fare"]-ks2["fare"].mean()) / ks2["fare"].std()
ks2.head()
```

```python
# Log Transformation Method
ks= sns.load_dataset("titanic")
ks.head()
```

```python
ks["fare"] = np.log(ks["fare"])
ks.head()
```

# Binning

- Grouping of values into smaller no of values
- Convert numeric into categories(bachay, jawan , bhooray)
- To have better understanding of groups

```python
kashti = sns.load_dataset("titanic")
bins = np.linspace(min(kashti["age"]), max(kashti["age"]), 15000)
age_groups = ["bachay","jawan","bhooray"]
kashti["age"] = pd.cut(kashti["age"], bins, labels=age_groups, include_lowest=True)
kashti["age"]
```
